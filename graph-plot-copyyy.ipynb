# (1 + 2) Vary Nodes + Vary flow 2-in-1 

```python
file_names = ['taskA-exp-nWifi-3','taskA-exp-nWifi-6','taskA-exp-nWifi-10','taskA-exp-nWifi-13','taskA-exp-nWifi-18']
for file_name in file_names:
    with open(file_name+'/throughput.dat') as file:
        lines = file.readlines()
    X = []
    Y = []
    for line in lines:
        words = line.split()
        words[0] = words[0].replace("ns","")
        words[0] = float(words[0])/(1e9)
        words[1] = float(words[1])
        X.append(words[0])
        Y.append(words[1])
        # print(words)
    # print(X)
    # print(Y)
    plt.rcParams["figure.figsize"] = (18,5)
    plt.xlabel('time (second)')
    plt.ylabel('throughput (Mbit/s)')
    plt.title('Network Throughput')
    plt.plot(X,Y,label=file_name)
    plt.legend()
```
