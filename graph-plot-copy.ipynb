```python
import matplotlib.pyplot as plt
```

```python
file_names = ['taskA-exp-nWifi-3','taskA-exp-nWifi-6','taskA-exp-nWifi-10']
for file_name in file_names:
    with open(file_name+'/throughput.dat') as file:
        lines = file.readlines()
    X = []
    Y = []
    for line in lines:
        words = line.split()
        words[0] = words[0].replace("ns","")
        words[0] = float(words[0])/(1e9)
        words[1] = float(words[1])
        X.append(words[0])
        Y.append(words[1])
        # print(words)
    # print(X)
    # print(Y)
    plt.rcParams["figure.figsize"] = (18,5)
    plt.xlabel('time (second)')
    plt.ylabel('throughput (Mbit/s)')
    plt.title('Network Throughput')
    plt.plot(X,Y,label=file_name)
    plt.legend()
```

```python
with open('wireless-sred-queue.dat') as file:
    lines = file.readlines()

X = []
Y = []
for line in lines:
    words = line.split()
    words[0] = float(words[0])
    words[1] = float(words[1])
    X.append(words[0])
    Y.append(words[1])
    # print(words)
# print(X)
# print(Y)

plt.rcParams["figure.figsize"] = (18,5)
plt.xlabel('time (second)')
plt.ylabel('buffer occupancy in bytes')
plt.title('Buffer Occupancy')

plt.ylim(0,500000)
plt.plot(X,Y)
```
